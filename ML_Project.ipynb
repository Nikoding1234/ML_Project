```python
x=iris.drop(['species'], axis =1)
```

```python
x
```

```python
Y=iris['species']
```

```python
Y
```

```python
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

Add random_state to the train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, y_train, y_test = train_test_split(x, Y, test_size = 0.4)
```

```python
xtrain, xtest, y_train, y_test = train_test_split(x, Y, test_size = 0.4, random_state=1)
```

**Logistic Regression**

```python
lr=LogisticRegression(solver= 'newton-cg')
```

```python
lr.fit(xtrain, y_train)
```

```python
y_pred1=lr.predict(xtest)
```

Get the confusion matrix

```python
confusion_matrix(y_test, y_pred1)
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_pred1), annot=True, cmap= 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

**Decision Tree**

```python
dtree=DecisionTreeClassifier()
dtree.fit(xtrain,y_train)
```

```python
y_pred2 = dtree.predict(xtest)
```

Get the confusion matrix, heatmap for confusion matrix

```python
confusion_matrix(y_test, y_pred2)
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_pred2), annot=True, cmap= 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
plt.figure(figsize=(10,10))
dec_tree = plot_tree(decision_tree=dtree, feature_names= iris.columns[:-1],
                     class_names =["sentosa", "vercicolor", "verginica"], filled = True )
```
